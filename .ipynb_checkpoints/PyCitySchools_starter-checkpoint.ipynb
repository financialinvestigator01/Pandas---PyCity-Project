# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
# [39170 rows x 11 columns]>
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
school_count = len(school_data_complete["school_name"].unique())

#Calculate the total number of students
total_students = school_data_complete["student_name"].count()

#Calculate the total budget
total_budget = school_data_complete.groupby(['school_name'])["budget"].unique().sum()

# Calculate the average math score
avg_math = round( (school_data_complete['math_score'].sum() / total_students), 6)

# Calculate the average reading score
avg_reading = round( (school_data_complete['reading_score'].sum() / total_students), 6 )

# Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_passing_rate = round ((avg_math + avg_reading) /2, 6)

# Calculate the percentage of students with a passing math score (70 or greater)
math_70 = [ mathscore for mathscore in school_data_complete['math_score'] if mathscore >= 70]
passing_math = round (len(math_70) / total_students * 100, 6)

# Calculate the percentage of students with a passing reading score (70 or greater)
reading_70 = [ readscore for readscore in school_data_complete['reading_score'] if readscore >= 70]
passing_reading = round  (len(reading_70) / total_students * 100, 6)
```

```python
# Create a dataframe to hold the above results
district_summary_df = pd.DataFrame({
                                    "Total Schools" : school_count,
                                    "Total Students" : total_students,
                                    "Total Budget" : total_budget,
                                    "Average Math Score" : avg_math,
                                    "Average Reading Score" : avg_reading,
                                    "% Passing Math" : passing_math,
                                    "% Passing Reading" : passing_reading,
                                    "% Overall Passing Rate" : overall_passing_rate        
                                    })
district_summary_df['Total Students'] = district_summary_df['Total Students'].map("{:,}".format)
district_summary_df['Total Budget'] = district_summary_df['Total Budget'].map("$ {:,}".format)
district_summary_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#initial list of schools names and types 
schoolname_group_by_df = school_data_complete.groupby(['school_name'])

#School Type
school_type = schoolname_group_by_df["type"].first()

#Total Students 
student_count = schoolname_group_by_df["student_name"].count()

#Total School Budget 
total_budget_per_school = schoolname_group_by_df["budget"].mean()

#Per Student Budget
school_budget_per_student = schoolname_group_by_df["budget"].mean()/schoolname_group_by_df["student_name"].count()

#Average Math Score
mean_math_score = schoolname_group_by_df["math_score"].mean()

#Average Reading Score
mean_reading_score = schoolname_group_by_df["reading_score"].mean()

#% Passing Math
passing_math_total_df = school_data_complete.loc[(school_data_complete["math_score"]>=70),:]
passing_math_total_groupby = passing_math_total_df.groupby(['school_name'])
passing_math = passing_math_total_groupby["student_name"].count()/schoolname_group_by_df["student_name"].count() * 100

#% Passing Reading
passing_reading_total_df = school_data_complete.loc[(school_data_complete["reading_score"]>=70),:]
passing_reading_total_groupby = passing_reading_total_df.groupby(['school_name'])
passing_reading = passing_reading_total_groupby["student_name"].count()/schoolname_group_by_df["student_name"].count() * 100

#% Overall Passing Rate
overall_passing_rate_by_school = (passing_math + passing_reading) /2 
```

```python
#    Create a dataframe to hold the above results

school_summary_df = pd.DataFrame({"School Type" : school_type,
                                    "Total Students" : student_count,
                                    "Total School Budget" : total_budget_per_school,
                                    "Per Student Budget" : school_budget_per_student,
                                    "Average Math Score" : mean_math_score,
                                    "Average Reading Score" : mean_reading_score,
                                    "% Passing Math" : passing_math,
                                    "% Passing Reading" : passing_reading,
                                    "% Overall Passing Rate" : overall_passing_rate_by_school
                                    })


#school_summary_formatted_df['Total Students'] = school_summary_df['Total Students'].map("{:,}".format)
school_summary_df['Total School Budget'] = school_summary_df['Total School Budget'].map("$ {:,}".format)
#school_summary_formatted_df['Per Student Budget'] = school_summary_df['Per Student Budget'].map("$ {:,.0f}".format)

# Top Performing Schools (By Passing Rate)
# Sort and display the top five schools in overall passing rate
school_summary_df.sort_values('% Overall Passing Rate', ascending=False)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
avg_scores_by_high_school_df = school_data_complete.groupby(['school_name', 'grade']).mean().reset_index()
avg_scores_by_high_school_df

# Create a pandas series for each grade. Hint: use a conditional statement.
ninth_grade_df = avg_scores_by_high_school_df.loc[avg_scores_by_high_school_df['grade'] == '9th']
tenth_grade_df = avg_scores_by_high_school_df.loc[avg_scores_by_high_school_df['grade'] == '10th']
eleventh_grade_df = avg_scores_by_high_school_df.loc[avg_scores_by_high_school_df['grade'] == '11th']
twelfth_grade_df = avg_scores_by_high_school_df.loc[avg_scores_by_high_school_df['grade'] == '12th']
twelfth_grade_df

# Group each series by school
ninth_df = pd.DataFrame({'School' : ninth_grade_df['school_name'], 
                         '9th Grade Math Score' : ninth_grade_df['math_score'],
                         '9th Grade Reading Score' : ninth_grade_df['reading_score']
                        })
tenth_df = pd.DataFrame({'School' : tenth_grade_df['school_name'], 
                         '10th Grade Math Score' : tenth_grade_df['math_score'],
                         '10th Grade Reading Score' : tenth_grade_df['reading_score']
                        })
eleventh_df = pd.DataFrame({'School' : eleventh_grade_df['school_name'], 
                            '11th Grade Math Score' : eleventh_grade_df['math_score'],
                            '11th Grade Reading Score' : eleventh_grade_df['reading_score']
                           })
twelfth_df = pd.DataFrame({'School' : twelfth_grade_df['school_name'], 
                           '12th Grade Math Score' : twelfth_grade_df['math_score'],
                           '12th Grade Reading Score' : twelfth_grade_df['reading_score']
                          })

# Combine the series into a dataframe

schools_df = pd.merge(ninth_df, tenth_df, on="School")
schools_df = pd.merge(schools_df, eleventh_df, on="School")
schools_df = pd.merge(schools_df, twelfth_df, on="School")

#Optional: give the displayed data cleaner formatting

mathscore_by_schools = pd.DataFrame({"School" : schools_df['School'],
                                    "9th Grade Math Score" : schools_df['9th Grade Math Score'],
                                    "10th Grade Math Score" : schools_df['10th Grade Math Score'],
                                    "11th Grade Math Score" : schools_df['11th Grade Math Score'],
                                    "12th Grade Math Score" : schools_df['12th Grade Math Score']
                                    })
mathscore_by_schools = mathscore_by_schools.set_index('School')
mathscore_by_schools.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
#Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:

# Average Math Score
# Average Reading Score
# % Passing Math
# % Passing Reading
# Overall Passing Rate (Average of the above two)

school_summary_df['Spending Ranges (per Student)'] = pd.cut(school_summary_df['Per Student Budget'],
                                                          spending_bins, labels=group_names)

ranges_by_student_df = pd.DataFrame({'Spending Ranges (per Student)': school_summary_df['Spending Ranges (per Student)'],
                                  'Average Math Score': school_summary_df['Average Math Score'],
                                  'Average Reading Score': school_summary_df['Average Reading Score'],
                                  '% Passing Math': school_summary_df['% Passing Math'],
                                  '% Passing Reading': school_summary_df['% Passing Reading'],
                                  '% Overall Passing Rate': school_summary_df['% Overall Passing Rate']
                                  })

scores_by_spending_df = ranges_by_student_df.groupby('Spending Ranges (per Student)').mean()
scores_by_spending_df
```
